```python
import numpy as np
import pandas as pd
```

```python
s={
    'Student':['Abi','shalfi','Ramya','keerthi'],
    'Marks':[95,98,94,np.nan]
}
```

```python
df=pd.DataFrame(s)
print(df)
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.fillna(value = 'Filling value')
```

```python
df.fillna(value = '45')
```

```python
Company={
    'Emp_id':[100,101,102,103,104,105],
    'Emp_name':['Abi','Ramya','Keerthi','Priya','suji','Gayathri'],
    'Emp_salary':[40000,35000,55000,32000,32000,np.nan]
}
```

```python
df=pd.DataFrame(Company)
print(df)
```

```python
df.groupby('Emp_salary').sum()
```

```python
df.groupby('Emp_name').count()
```

```python
# byComp['Emp_salary'].mean()
mean_salary_by_name = df.groupby('Emp_name')['Emp_salary'].mean()
print(mean_salary_by_name)
```

```python
df.groupby('Emp_name').describe()
```

```python
std_salary_by_name = df.groupby('Emp_name')['Emp_salary'].std()
print(std_salary_by_name)
```
