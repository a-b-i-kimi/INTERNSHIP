```python
a= df['price']
b=df['area']
plt.scatter(a,b)
```

```python
plt.figure(figsize=(12,6))
plt.plot(a,b)
plt.show()
```

```python
area = df['area'].astype(str)
price = df['price']
plt.figure(figsize=(15,6))
plt.bar(area,price)
plt.xticks(ticks=range(0, len(area), 12), labels=area[::12], rotation=90)
#plt.yticks(np.arange(0,df['price'].max()))
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.tight_layout()
plt.show()
```

```python
area = df['area'].astype(str)
price = df['price'].astype(str)
plt.figure(figsize=(10,20))
plt.barh(area,price)
plt.xticks(ticks=range(0, len(area), 10), labels=area[::10], rotation=90)
plt.yticks(ticks=range(0, len(price), 10), labels=price[::10],rotation=0)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.tight_layout()
plt.show()
```

```python
topn=df.sort_values(by='price',ascending=False).head(10)
#print(topn)
plt.figure(figsize=(10,6))
plt.xticks(rotation=45)
sns.barplot(x='area',y='price',data=topn)
```

```python
x=df['price']
plt.figure(figsize=(12,6))
plt.hist(x,color='green',edgecolor='black',bins=20)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
x=df['price']
plt.figure(figsize=(20,6))
sns.displot(x,color='red',edgecolor='black',bins=20,kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
